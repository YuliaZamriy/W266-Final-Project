## Aggregating all files

```python
keep_cols_descr = ['speech_id', 'date', 'char_count', 'word_count', 'speaker']
keep_cols_descr_excl = ['speech_id', 'date', 'char_count', 'word_count', 'speaker', 'state', 'gender', 'first_name', 'last_name']
keep_cols_map = ['speakerid', 'speech_id', 'Party', 'Congress', 'Chamber', 'State', 'Ethnicity', 'Age', 'Age_lt_med', 'NonWhite', 'Female', 'Full Name']
```

```python
congresses = []
for c in range(97,115):
    c = str(c)
    if len(c)==2:
        c = '0'+c
    congresses.append(c)
congresses
```

```python
ls ../raw
```

```python
dirpath = '../raw/hein-daily/'
full_descr = pd.DataFrame()
full_descr_excl = pd.DataFrame()
row_cnt = 0
for c in congresses:
    mapfile = dirpath+c+'_SpeakerMap.txt'
    descrfile = dirpath+'descr_'+c+'.txt'
    speakermap = pd.read_csv(mapfile, sep="|")
    speakermap = speakermap.merge(full_map, how='left', on='speakerid')
    descr = pd.read_csv(descrfile, sep="|")
    descr_exp = descr[keep_cols_descr].merge(speakermap[keep_cols_map], how='inner', on='speech_id')
    descr_excl = descr[keep_cols_descr_excl].merge(speakermap[['speech_id','speakerid']], how='left', on='speech_id')
    descr_excl = descr_excl[descr_excl.speakerid.isna()]
    row_cnt += descr_exp.shape[0]
    print("File: {} raw: {} rows. Processed: {} rows".format(c, descr.shape[0], descr_exp.shape[0]))
    full_descr = full_descr.append(descr_exp, ignore_index=True)
    full_descr_excl = full_descr_excl.append(descr_excl, ignore_index=True)
```

```python
full_descr.shape
```

```python
row_cnt
```

```python
full_map['Congress'] = full_map['Congress'].astype(str)
full_map['Congress'].describe()
```

```python
temp = full_descr_excl[(full_descr_excl.last_name != 'Unknown') & (full_descr_excl.gender.isin(['F','M']))]
temp.shape
```

```python
gender_match = temp.merge(full_map, how='left', left_on=['last_name','gender','Congress'], right_on=['last_name','Gender','Congress'])
gender_match = gender_match[~gender_match['Ethnicity'].isna()]
gender_match.shape
```
